```python
a=women['long_hair'].mean()
b=women['nose_wide'].mean()
c=women['nose_long'].mean()
d=women['lips_thin'].mean()
e=women['distance_nose_to_lip_long'].mean()
f=man['long_hair'].mean()
g=man['nose_wide'].mean()
h=man['nose_long'].mean()
t=man['lips_thin'].mean()
k=man['distance_nose_to_lip_long'].mean()
z=women['forehead_width_cm'].mean()
x=women['forehead_height_cm'].mean()
o=man['forehead_width_cm'].mean()
p=man['forehead_height_cm'].mean()
```

```python
p
```

```python
wo=[a,b,c,d,e]
ma=[f,g,h,t,k]
X=['long_hair','nose_wide','nose_long','lips_thin','distance_nose_to_lip_long']
color1 = '#B35E5E'
color2 = '#5E74B3'

w = 0.2

bar1 = np.arange(len(X))
bar2 = [i + w for i in bar1] 

plt.figure(figsize=(10,10), dpi= 80)

plt.bar(bar1, wo, w, label = 'women', color = color1)
plt.bar(bar2, ma, w, label = 'man',color = color2)

plt.xlabel("CATEGORIES")
plt.ylabel("MEAN VALUES")
plt.title("COMPARISON")
plt.xticks(bar1+0.1,X)

plt.legend()

plt.show()
```

```python
fe=[z,x]
male=[o,p]
labelss=['forehead_width_cm','forehead_height_cm']

color1 = '#B35E5E'
color2 = '#5E74B3'

l = 0.2

bar3 = np.arange(len(labelss))
bar4 = [i + l for i in bar3] 

plt.figure(figsize=(10,10), dpi= 80)

plt.bar(bar3, fe, l, label = 'women', color = color1)
plt.bar(bar4, male, l, label = 'man',color = color2)

plt.xlabel("CATEGORIES")
plt.ylabel("MEAN VALUES")
plt.title("COMPARISON")
plt.xticks(bar3+0.1,labelss)

plt.legend()

plt.show()
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
df = data.copy()
  
# calling the get_dummies method
# the first parameter mentions the
# the name of the data frame to store the
# new data frame in
# the second parameter is the list of
# columns which if not mentioned
# returns the dummies for all
# categorical columns
df = pd.get_dummies(df,
                     columns = ['gender'])
  
display(df)
```

```python
df1=df.drop('gender_Male',axis=1)
```

```python
df1
```

```python
plt.rcParams["figure.figsize"]=(10,6)

plt.subplot(2,2,1)
plt.scatter(df1['forehead_height_cm'],df1['gender_Female'],c='r')
plt.grid()

plt.subplot(2,2,2)
plt.scatter(df1['forehead_width_cm'],df1['gender_Female'])
plt.grid()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1[['long_hair',
                                                         'forehead_width_cm',
                                                         'forehead_height_cm',
                                                         'nose_wide',
                                                        'nose_long',
                                                        'lips_thin',
                                                        'distance_nose_to_lip_long']],df1.gender_Female,train_size=0.8)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_predicted = model.predict(X_test)
```

```python
y_predicted
```

```python
model.score(X_test,y_test)
```
